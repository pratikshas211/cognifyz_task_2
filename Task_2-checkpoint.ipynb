```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import LabelEncoder
```

```python
# Load dataset
df=pd.read_csv("C:/Users/prati/Downloads/Dataset .csv")
```

```python
# Display basic info
display(df.head())
display(df.info())
```

```python
df_cleaned = df[['Restaurant Name', 'City', 'Cuisines', 'Price range', 'Aggregate rating', 'Votes']].copy()
df_cleaned.dropna(subset=['Cuisines', 'Price range', 'Aggregate rating'], inplace=True)
```

```python
# Convert cuisines to lowercase for consistency
df_cleaned['Cuisines'] = df_cleaned['Cuisines'].str.lower()
```

```python
# Encoding categorical features
label_enc = LabelEncoder()
df_cleaned['Price range'] = label_enc.fit_transform(df_cleaned['Price range'])
```

```python
# Convert cuisines into numerical features using TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words='english')
cuisine_matrix = vectorizer.fit_transform(df_cleaned['Cuisines'])
```

```python
# Compute similarity using Cosine Similarity
cosine_sim = cosine_similarity(cuisine_matrix, cuisine_matrix)
```

```python
# Recommendation Function
def recommend_restaurants(user_cuisine, user_price_range, top_n=5):
    # Filter restaurants by price range
    filtered_df = df_cleaned[df_cleaned['Price range'] == user_price_range]

    if filtered_df.empty:
        return "No restaurants found for the given price range."
    
    # Find the first restaurant index matching the user's cuisine
    matching_index = filtered_df[df_cleaned['Cuisines'].str.contains(user_cuisine, case=False, na=False)].index

    if matching_index.empty:
        return "No matching restaurants found for the selected cuisine."

    # Use the first matching restaurant as a reference
    reference_idx = matching_index[0]

    # Compute similarity scores
    sim_scores = list(enumerate(cosine_sim[reference_idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]

    # Get recommended restaurant indices
    recommended_indices = [i[0] for i in sim_scores]

    return df_cleaned.iloc[recommended_indices][['Restaurant Name', 'City', 'Cuisines', 'Aggregate rating', 'Votes']]
        
```

```python
# Example Test Case
print(recommend_restaurants('japanese', 2))
```
